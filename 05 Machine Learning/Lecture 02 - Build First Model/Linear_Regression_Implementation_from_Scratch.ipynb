## Linear Regression Implementation from Scratch

#### Step1 : Prepare the dataset

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset
x = np.array([1, 2, 3, 4, 5])  # Feature
y = np.array([3, 5, 7, 9, 11])  # Target (follows y = 2x + 1)

# Visualize the dataset
plt.scatter(x, y, color="blue", label="Data Points")
plt.title("Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

#### Step 2: Define the Model

```python
# Linear regression model
def predict(x, theta0, theta1):
    """
    Predicts the value of y given x, theta0, and theta1.
    """
    return theta0 + theta1 * x

# Initialize parameters (starting values)
theta0 = 0  # Intercept
theta1 = 0  # Slope

# Example prediction
predicted_y = predict(x, theta0, theta1)
print("Predicted values with initial theta0 = 0, theta1 = 0:", predicted_y)
```
