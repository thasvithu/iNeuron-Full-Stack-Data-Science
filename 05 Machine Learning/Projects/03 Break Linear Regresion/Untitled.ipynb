## Breaking Down Linear Regression from Scratch

#### Why Do We Need Linear Regression?

Let’s begin by understanding why Linear Regression is useful in real-world scenarios. Imagine we have a dataset with two attributes: "Study Hours" (X) and "Scores" (Y).

Here's a simplified version of the data:

x = [1, 2, 3, 4]

y = [3, 5, 7, 9]



*Note*

In this example, we’re using a small dataset for simplicity. Working with a smaller dataset helps focus on the core concept of Linear Regression. In real-world scenarios, datasets are typically larger and more complex, and drawing conclusions can become more challenging.

Now, let's visualize this dataset on a graph using the matplotlib library in Python:

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4]
y = [3, 5, 7, 9]

# Plotting the data points
plt.plot(x , y, marker='o')
plt.xlabel("Study Hours (X)")
plt.ylabel("Scores (Y)")
plt.title("Study Hours vs Scores")
plt.show()
```

When you plot the data, you will see that all the points align perfectly along a straight line. This indicates a linear relationship between the "Study Hours" and "Scores". However, keep in mind that in real-world data, perfect alignment like this is rare. Usually, data points will vary, resulting in some errors.

###### The Equation of a Line: y = mx + c
A line can be represented mathematically by the equation:

$
    y = mx + c
$

Where:
- m is the slope of the line (how steep it is).
- c is the intercept (where the line crosses the y-axis).


Let’s calculate the slope (m) and intercept (c) manually using the points in our dataset.

We'll use the points (x1, y1) = (1, 3) and (x2, y2) = (2, 5) to calculate the slope and intercept.

###### Step 1: Calculate the Slope (m)
The formula for the slope between two points is:

m = (y2 - y1) / (x2 - x1)

```python
x1 = 1
y1 = 3
x2 = 2
y2 = 5

#lets find the slope
m = (y2 - y1) / (x2 - x1)
print(f"Slope(m) is : {m}")
```

###### Step 2: Calculate the Intercept (c)
To find the intercept c, we use the equation y = mx + c. Let's plug in the values of x1 = 1 and y1 = 3

```python
#lets find the intercept using x1, y1
#y = mx + c
#c = y - m * x

x1 = 1
y1 = 3

c = y1 - m * x1
print(f"Intercept(c) is : {c}")
```

###### Step 3: Making Predictions
Now that we have the equation of the line, we can use it to predict the values of y for any given x. 

For example:
    
Let’s predict the scores for study hours of 10, 15, and 50:

```python
x_new = [10, 15, 50]

for i in x_new:
    y_pred = 2 * i + 1
    print(f"When study hours = {i}, the predicted score is: {y_pred}")
```

###### What is a Model?
You might be wondering: What exactly is a model?

Don’t worry, the term model can sometimes seem complex, but in this context, it’s simply a mathematical equation. In our case, the equation y=2x+1 is the model. Using this equation, we can predict the output (scores) for any input (study hours).

###### What is the Mean Error?
It’s crucial to understand error in this context. Let’s take the first two points from our dataset:

x1 = 1 and y1 = 3

After building our model, if you input x = 1, the model will predic y = 3. In this case, there’s no error because the model predicted the actual value correctly.

However, if the model predicts a value that differs from the actual value, that’s an error. And while a model might predict the correct value for some inputs, in the real world, it rarely happens consistently. In most cases, there will be some error in the predictions.

###### What Happens in the Real World?
In real-world scenarios, datasets aren’t as neat and perfectly aligned as the one we just used. When you plot real-world data, you’ll notice that the points won’t fall on a straight line. The data might have some "noise", meaning there will be discrepancies between the actual values and the predicted values.

Because of this, predictions using a perfect line would likely be inaccurate. This is where Linear Regression becomes essential!

###### How Linear Regression Helps
Linear Regression helps by finding the best-fit line, which minimizes the errors between the predicted values and the actual data points. The algorithm will adjust the slope (m) and intercept (c) in such a way that the total error (difference between the actual and predicted values) is minimized.

In future sections, we will explore techniques like Gradient Descent to optimize the model parameters (m and c) and minimize the error.

#### What is Linear Regression?
Linear Regression is a supervised machine learning algorithm used to make predictions when there is a linear relationship between the variables in the dataset. In other words, if the data shows a linear trend, Linear Regression can be used to model and predict the outcomes.

##### Mean Squared Error (MSE)
MSE is commonly used to find the error in regression tasks.

The equation for MSE is:

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
- MSE: Mean Squared Error
- n: Number of data points
- yi: Actual value of the i-th data point
- y^i : Predicted value of the i-th data point

Let’s calculate the total error when m = 0 and c = 0:

```python
import numpy as np

# Actual values
y = [2, 6, 7, 12]

# Predicted values (from m = 0, c = 0)
y_pred = [0, 0, 0, 0]

# Function to calculate MSE
def find_mse(n, y_act, y_pred):
    return np.sum((np.array(y_act) - np.array(y_pred)) ** 2) / n

# Calculating MSE
mse = find_mse(4, y, y_pred)
print(f"MSE is: {mse}")
```

Now we’ve calculated the total error—58.25. In Linear Regression, the main aim is to reduce the error towards 0. But how do we update m and c to minimize the error?

#### Gradient Descent: The Optimization Algorithm

In Linear Regression, the goal is to minimize the error (cost function) by finding the best values for m and c. Gradient Descent is an optimization algorithm that adjusts m and c iteratively to reduce the error.

The process works as follows:
1. Start with initial values for m and c (e.g., 0).
2. Calculate the error (Mean Squared Error) for the current values.
3. Compute the gradients (∂J/∂m and ∂J/∂c) to determine the direction to move.
4. Update m and c using the formulas

m = m - α . ∂J/∂m

c = c - α . ∂J/∂c

where α is the learning rate that controls how much m and c change in each iteration

5. Repeat steps 2–4 until the error converges to a minimum value.

By iteratively updating m and c, Gradient Descent ensures that the model finds the optimal line that minimizes the error.

Lest plot the Error(MSE) vs m for the first error

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
x = np.array([1, 2, 3, 4])
y = np.array([2, 6, 7, 12])

# Define a range of 'm' values
m_values = np.linspace(-2, 4, 100)  # Generate 100 points between -2 and 4
c = 0  # Keep intercept constant
mse_values = []  # To store the computed MSE for each 'm'

# Function to calculate Mean Squared Error
def calculate_mse(y_actual, y_pred):
    return np.mean((y_actual - y_pred) ** 2)

# Loop through each 'm' and calculate MSE
for m in m_values:
    y_pred = m * x + c  # Predict using y = mx + c
    mse = calculate_mse(y, y_pred)  # Calculate MSE
    mse_values.append(mse)

# Plot Error (MSE) vs. Slope (m)
plt.figure(figsize=(8, 6))
plt.plot(m_values, mse_values, label="MSE vs. m", color="blue")
plt.xlabel("Slope (m)")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Error (MSE) vs. Slope (m)")
plt.axvline(x=0, color='red', linestyle='--', label="Initial m=0")
plt.legend()
plt.grid()
plt.show()
```
