## Polynimial Regression implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 6 * np.random.rand(100, 1) - 3
y = 0.5 * x**2 + 1.5 * x + 2 + np.random.randn(100, 1)
# this is my quadratic equation used = 0.5x^2 + 1.5x+2 + outliers
```

```python
# lest plot x,y
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("y")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# lest implement simple linear regresssion
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, linear_model.predict(X_test))
score
```

```python
## lest visualize this model
plt.plot(X_train, linear_model.predict(X_train), c="r")
plt.scatter(X_train, y_train)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# lest apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
linear_model.fit(X_train_poly, y_train)
y_pred = linear_model.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```
