#### First we will build a model TV vs Sales

```python
x = df[["TV"]]
x
```

```python
y = df["Sales"]
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
#in this step our model is done which mean we found out the m and c
```

```python
#so now how to the m and c values
model.intercept_ #this is our c value
```

```python
model.coef_ #this is our m value
```

```python
#now our equation is sales = m.TV + C
#sales = 0.047536TV + 7.03259
```

```python
#we can save the mnodel for future purpose or we can share with others
file = "firt_model.sav"
pickle.dump(model, open(file, "wb"))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Initial guess for m and c
m = 0
c = 0

# Learning rate
alpha = 0.01

# Function to calculate predictions
def predict(x, m, c):
    return m * x + c

# Function to update m and c using gradient descent
def update_parameters(x, y, m, c, alpha):
    N = len(y)
    m_gradient = -2/N * sum(x * (y - (m*x + c)))
    c_gradient = -2/N * sum(y - (m*x + c))
    m = m - alpha * m_gradient
    c = c - alpha * c_gradient
    return m, c

# Plotting
plt.figure(figsize=(10,6))

# Iteratively update and plot the line
for i in range(100):
    # Update m and c
    m, c = update_parameters(x, y, m, c, alpha)
    
    # Plot the data points
    plt.scatter(x, y, color='blue')
    
    # Plot the current line
    plt.plot(x, predict(x, m, c), color='red')
    
    # Display the current line
    plt.title(f'Iteration {i+1}: m={m:.2f}, c={c:.2f}')
    plt.pause(0.1)  # Pause to visualize the change

plt.show()
```
