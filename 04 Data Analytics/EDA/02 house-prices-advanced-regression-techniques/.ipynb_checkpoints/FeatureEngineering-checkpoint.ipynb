### Handling missing values

```python
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 0]
for feature in features_nan:
    print(f"The feature is {feature} and missing values is {np.round(dataset[feature].isnull().mean() * 100, 4)}%")
```

```python
## Replacing categorical features
categorical_features = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
categorical_features
```

```python
#method 1 with using a function
data = dataset.copy()
def replace_cat_features(dataset, categorical_features):
    data[categorical_features] = data[categorical_features].fillna('Missing')
    return data
```

```python
replace_cat_features(dataset, categorical_features)
```

```python
# Method 2: Using Mode
data = dataset.copy()

def replace_using_mode(dataset, categorical_features):
    for feature in categorical_features:
        # Fill missing values in each categorical feature with its mode
        dataset[feature].fillna(dataset[feature].mode()[0], inplace=True)
    return dataset

data = replace_using_mode(data, categorical_features)
```

```python
replace_using_mode(dataset, categorical_features)
```

```python
numerical_features_with_nan = [feature for feature in dataset.columns if dataset[feature].dtypes != 'object'
                                and dataset[feature].isnull().sum() >= 1]
numerical_features_with_nan
```

```python
for feature in numerical_features_with_nan:
    ##since there are outliers we are going to replace with median
    median_values = dataset[feature].median()
    
    ##create a new feature to capture nan value
    dataset[feature + 'nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature].fillna(median_values, inplace = True)
```
